## Loading and exploring the dataset (6pts)

Let's start by loading the dataset. you'll have to use the `fetch_20newsgroups` function from [`sklearn.datasets`](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.fetch_20newsgroups.html):


```python
from sklearn.datasets import fetch_20newsgroups

#---------------------------------------------------------------------------------------
# Load the training section of the dataset
newsgroups_data = fetch_20newsgroups(subset='train')

# Explore the loaded data

#---------------------------------------------------------------------------------------

# print number of the training articles
# YOUR CODE HERE
no_of_training_articles = len(newsgroups_data.data)
print("The number of training articles in tha dataset: \n",no_of_training_articles)

#---------------------------------------------------------------------------------------

# print number of categories
# YOUR CODE HERE
categories_of_data = newsgroups_data.target_names
no_of_categories =  len(categories_of_data)
print("The number of categories in the dataset: \n",no_of_categories)

#---------------------------------------------------------------------------------------

# print first 5 category names
# YOUR CODE HERE
print("The first 5 categories in the dataset:")
for i in range(5):
    print(categories_of_data[i])
```

## visualizing number of examples in each category (4 pts)

Create a barplot using `seaborn` or `matplotlib` library to show the number of examples in each newsgroup
x-axis should show the newsgroup title and y-axis the number of articles.

Both [`seaborn`](https://seaborn.pydata.org/) and [`matplotlib`](https://matplotlib.org/stable/) are popular libraries for visualization in python. You can use any of them to complete this task.

```python
# YOUR CODE HERE

import matplotlib.pyplot as plt
import numpy as np

#-----------------------------------------------------------------------------------------------------------

## In order to complete the bar plot we need to extract the number of articles for each newsgroup category

## Extracted the names of categories in categories_of_data and the number of categories in no_of_categories,
## using these variables we can plot the bar graph

category_index = newsgroups_data.target
## Here category index contains a list of all the the indices associated with the category like for example 
## 0 for category1, 1 for category2

## We need to iterate over the category index to find the number of example in each category

category_no_of_articles={}
for index in category_index:
    category_name = categories_of_data[index]
    if category_name in category_no_of_articles:
        category_no_of_articles[category_name] += 1
    else:
        category_no_of_articles[category_name] = 1    
print(category_no_of_articles)

#--------------------------------------------------------------------------------------------------------------

## Plot the bar plot
categories_plot = list(category_no_of_articles.keys())
no_example_plot = list(category_no_of_articles.values())

cmap = plt.get_cmap('tab20')
plt.figure(figsize=(8, 6))
plt.bar(categories_plot, no_example_plot, color=cmap(np.arange(len(categories_plot))))
plt.xlabel('Category of Newsgroup')
plt.ylabel('Number of Articles')
plt.title('Number of Articles in Each Newsgroup Category')
plt.xticks(rotation=45, ha='right', fontsize=12)  # Rotate x-axis labels for better readability
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add grid lines
plt.tight_layout()

plt.show()
```

# Featurizing Text (10pts)

To use textual data for machine learning models, we need to convert the text into numerical values. In this section we'll explore a basic approach for converting a document into numerical vecotors called `Bag of Words (BoW)`

## Bag of Words (BoW)
Bag of Words is a simple yet effective technique in text processing. It involves representing text data by counting the frequency of each word in a document, disregarding grammar and word order.
For more information about what a BoW representation is you can refer to the beginning of [this video](https://www.youtube.com/watch?v=nZromH6F7R0)

Load the training section of the dataset and using sklearn's [`CountVectorizer`](https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.CountVectorizer.html) build a count vectorizer over it. **Your vectorizer should take 1-grams and 2-grams as feature and remove english stop words. Also it should only take n-grams which at most appear in 80% of the documents and at least appear in 0.1% of documents.**

Then transform both training and test datasets
to to bag of word representations and write out the shape of both training and test datasets:

**shape of a numpy array x can be found using** `x.shape`



```python
from sklearn.feature_extraction.text import CountVectorizer

#------------------------------------------------------------------------------------------------------------------------

# Load training and test portions of the dataset
# YOUR CODE HERE
train_data = fetch_20newsgroups(subset='train')
test_data = fetch_20newsgroups(subset='test')

#------------------------------------------------------------------------------------------------------------------------

# Initialize the CountVectorizer with required parameters
# YOUR CODE HERE
cnt_vectorizer = CountVectorizer(ngram_range=(1,2),stop_words='english',min_df=0.001,max_df=0.8)

## Here while initializing CountVectorizer ngram_range=(1,2) specifies that it can take both 1-grams and 2-grams as 
## features. Along with that we need to remove english stop words ; and as n-grams which at most appear in 80% of 
## document and at least appear 0.1% of documents; min_df = 0.001 and max_df = 0.8 as min_df specifies minimum number
## of n-grams to considered and max_df specifices maximum number of n-grams to be considered

#---------------------------------------------------------------------------------------------------------------------

# Build count vectors from the dataset
# YOUR CODE HERE
train_data_bow = cnt_vectorizer.fit_transform(train_data.data)

## Using fit_transform on train data ensures to learn the vocaboulary from train data and transform the train data 
## into matrix of token counts.In short it first gets trained on vocabulary and then transform it into token counts.

test_data_bow = cnt_vectorizer.transform(test_data.data)

## In the train data its just needed to transform the data into token counts and no learning process is required from
## the test data. Thus using transform for test data.

#----------------------------------------------------------------------------------------------------------------------

# print shape of the trainig and test datasets
# YOUR CODE HERE
print("Shape of training dataset : ",train_data_bow.shape)
print("Shape of test dataset : ",test_data_bow.shape)
```

# Logistic Regression

## Training and Evaluating Logistic Regression Model on BoW features (30pts)

Logistic Regression is a foundational statistical method in machine learning for predicting a continuous variable. This section delves into training a Logistic Regression model (a linear model used for classification tasks), focusing on the '20 newsgroups' dataset. You'll learn to split the dataset, initialize the model, fit the model to your training data, and evaluate its performance on the test set. This hands-on experience will solidify your understanding of linear models in a classification context.

You are supposed to use [sklearn.linear_model.LogisticRegressionCV](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegressionCV.html) For this section. 'LogisticRegressionCV' is a logistic regression model with built-in cross validation capability. Read the document carefully and implement the following steps:

- Define a Logistic Regression model with L2 regularization that runs for 100 iterations and 4-fold cross validation
- Train the model over the BoW features of the training data from previous step
- Measure accuracy, precision, recall and F1 score of the trained model over test dataset

For measuring the metrics you can use `sklearn` library as well. Note that since we have a non-binary classification we need to average precision, recall and f1 metrics. For averaging use `macro` aggregation.

```python
%%time 

from sklearn.linear_model import LogisticRegressionCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import warnings
warnings.filterwarnings("ignore")

#---------------------------------------------------------------------------------------------------------------------

# Initialize the Logistic Regression model
# YOUR CODE HERE
logistic_regression_model = LogisticRegressionCV(Cs=[0.01,0.1,1], cv=4, max_iter=100, penalty='l2', scoring='accuracy', random_state=42)

#---------------------------------------------------------------------------------------------------------------------

# Fit the model to the training data
# YOUR CODE HERE
logistic_regression_model.fit(train_data_bow, train_data.target)

#---------------------------------------------------------------------------------------------------------------------

# Predict the labels on the test set and measure metrics
# YOUR CODE HERE
test_pred_logistic = logistic_regression_model.predict(test_data_bow)

accuracy_logistic = accuracy_score(test_data.target, test_pred_logistic)
precision_logistic = precision_score(test_data.target, test_pred_logistic, average='macro')
recall_logistic = recall_score(test_data.target, test_pred_logistic, average='macro')
f1_logistic = f1_score(test_data.target, test_pred_logistic, average='macro')

#---------------------------------------------------------------------------------------------------------------------

# Print the metrics
print("Accuracy:", accuracy_logistic)
print("Precision:", precision_logistic)
print("Recall:", recall_logistic)
print("F1 Score:", f1_logistic)
```

## NOTE:
1. A variety of criteria, including training time and accuracy, must be taken into account while building the best logistic regression model,  making it necessary to tune the hyperparameters.
2. When building a logistic regression model, the regularization strength parameter (Cs) plays a crucial role in balancing model complexity and generalization performance. It determines the degree of regularization applied, with smaller values indicating stronger regularization and larger values indicating weaker regularization.
3. While experiments using both maximum and minimum values for the Cs parameter, the accuracy was around 80%
4. I have used the lower cs value since it was taking a lesser time to train

decision more than the others. Use the model trained on TF-IDF features. Analyze the coefficients of the trained model for each class and print top-10 features along with their coefficient. For example we might have the following list for the class `comp.sys.mac.hardware`:

```python
# YOUR CODE HERE
import numpy as np

#---------------------------------------------------------------------------------------------------------------------

# Get feature names from the CountVectorizer which was initalized above
feature_names = cnt_vectorizer.get_feature_names_out()

class_coefficients = {}

#---------------------------------------------------------------------------------------------------------------------

# Get coefficients for each newgroup category
for class_name in train_data.target_names:
    class_index = train_data.target_names.index(class_name)
    coefficients = logistic_regression_model.coef_[class_index]
    class_coefficients[class_name] = coefficients
    
#---------------------------------------------------------------------------------------------------------------------
    
# Print top-10 influential features for each newgroup category based on the coefficients
for class_name, coefficients in class_coefficients.items():
    print(f"Top 10 influential features for class {class_name} along with their coefficients:")
    sorted_features = sorted(zip(coefficients, range(len(coefficients))), key=lambda x: x[0], reverse=True)
    top_indices = [index for _, index in sorted_features[:10]]
    for index in top_indices:
        print(f"{feature_names[index]}: {coefficients[index]:.4f}")
    print()
```

# Training and Evaluating Simple Neural Network Model (30pts)

The Multi-Layer Perceptron (MLP) is a versatile algorithm in machine learning, capable of handling complex patterns in data through its layered structure of neurons. This section is dedicated to training a Multi-Layer Perceptron model using the '20 newsgroups' dataset, focusing on its application in classification tasks. You will go through the process of splitting the dataset, initializing the MLP model, fitting the model with your training data, and evaluating its performance on the test set. This practical exercise will enhance your understanding of neural networks in the realm of classification.

For this section, you are asked to utilize [sklearn.neural_network.MLPClassifier](https://scikit-learn.org/stable/modules/generated/sklearn.neural_network.MLPClassifier.html). 'MLPClassifier' is an implementation of a multi-layer perceptron classifier. Carefully review the documentation and carry out the following steps:

- Define a Multi-Layer Perceptron model with at two hidden layers of size 100 and `sigmoid` activation function. Use a batch size of 128 and train the model for 25 iterations.
- Train the MLP model using the Bag of Words (BoW) features of the training data obtained in the previous step.
- Assess the accuracy, precision, recall, and F1 score of the MLP model on the test dataset.

For the evaluation metrics, continue to use the `sklearn.metrics` module. Given the non-binary nature of the classification in this dataset, ensure to average the precision, recall, and F1 metrics using the 'macro' strategy to account for the imbalance across classes.


```python
%%time
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

#---------------------------------------------------------------------------------------------------------------------

# Initialize the Multi-Layer Perceptron model
multi_layer_perceptron_model = MLPClassifier(hidden_layer_sizes=(100, 100), activation='logistic', batch_size=128, max_iter=25, random_state=42)

#---------------------------------------------------------------------------------------------------------------------

# Fit the multi_layer_perceptron_model to the training data
multi_layer_perceptron_model.fit(train_data_bow, train_data.target)

#---------------------------------------------------------------------------------------------------------------------

# Predict the labels on the test  on multi_layer_perceptron_model
test_pred = multi_layer_perceptron_model.predict(test_data_bow)

#---------------------------------------------------------------------------------------------------------------------

# Measure metrics for multi_layer_perceptron_model
accuracy_multi_layer_perceptron_model = accuracy_score(test_data.target, test_pred)
precision_multi_layer_perceptron_model = precision_score(test_data.target, test_pred, average='macro')
recall_multi_layer_perceptron_model = recall_score(test_data.target, test_pred, average='macro')
f1_multi_layer_perceptron_model = f1_score(test_data.target, test_pred, average='macro')

#---------------------------------------------------------------------------------------------------------------------

# Print the metrics for multi_layer_perceptron_model
print("Accuracy:", accuracy_multi_layer_perceptron_model)
print("Precision:", precision_multi_layer_perceptron_model)
print("Recall:", recall_multi_layer_perceptron_model)
print("F1 Score:", f1_multi_layer_perceptron_model)
```

## NOTE :
1. The MLP model shows strong performance with high accuracy and quick training, indicating well-optimized hyperparameters. 
2. The chosen architecture and its many parameters, such as the number of layers, neuron size, activation functions, and batch size, appear to achieve a satisfactory balance between speed and complexity.
